# Creating voting affinity scores for members of the 118th Congress

```python
import numpy as np
import pandas as pd
```

```python
votes=pd.read_csv("https://voteview.com/static/data/out/votes/HS118_votes.csv")
```

```python
votes = votes[["chamber","rollnumber","icpsr","cast_code"]]
s_votes = votes.query("chamber == 'Senate'")
```

```python
paired_votes = pd.merge(s_votes, s_votes,
                       on = ['chamber','rollnumber'],
                       how = 'inner',
                       validate = 'many_to_many')
```

```python
paired_votes
```

```python
paired_votes = paired_votes.rename({'icpsr_x':'icpsr',
                                   'icpsr_y':'member_compare',
                                   'cast_code_x':'vote',
                                   'cast_code_y':'compare_vote'
                                   }, axis=1)
paired_votes = paired_votes.query('icpsr != member_compare')
paired_votes
```

```python
paired_votes['equal_vote']=paired_votes['vote']==paired_votes['compare_vote']
paired_votes

paired_votes = paired_votes.assign(equal_vote =paired_votes['vote']==paired_votes['compare_vote'])
```

```python
paired_vote = paired_votes[['icpsr','member_compare','equal_vote']]
paired_vote
```

```python
affinity = paired_votes.groupby(['icpsr','member_compare'])\
    .agg({'equal_vote':'mean'}).reset_index()
```

```python
affinity['affinity_score'] = round(100*affinity['equal_vote'], 1).astype('str')+'%'
```

```python
affinity = affinity.drop('equal_vote', axis=1)
```

```python
ideology = pd.read_csv("https://voteview.com/static/data/out/members/HS118_members.csv")
ideo_senate = ideology.query("chamber == 'Senate'")
```

```python
ideo_senate.columns
```

```python
ideo_senate = ideo_senate[['icpsr','state_abbrev','party_code','bioname']]
ideo_senate
```

```python
affinity = pd.merge(affinity,ideo_senate,
                    on = 'icpsr',
                   how = 'outer',
                   validate = 'many_to_one')
affinity
```

```python
affinity = pd.merge(affinity,ideo_senate,
                    left_on = 'member_compare',
                    right_on = 'icpsr',
                   how = 'outer',
                   validate = 'many_to_one')
affinity
```

```python
affinity.query("state_abbrev_x == 'VA'").sort_values("affinity_score")
```
